```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
year = ['2013', '2014', '2015', '2016']

# Cancer Deaths Avoided
data1 = {'year': year,
         'number_of_cancer_deaths_avoided': [52618,55237,60293,59981]}
df1 = DataFrame(data1, columns=['year','number_of_cancer_deaths_avoided'])
sns.lineplot(x="year", y="number_of_cancer_deaths_avoided", data=df1)
plt.show()
```

```python
# All Cancer Combined Incidence
data2 = {'year': year,
         'all_cancer_combined_incidence': [1264,1243,1230,1209]}
df2 = DataFrame(data2, columns=['year','all_cancer_combined_incidence'])
sns.lineplot(x="year", y="all_cancer_combined_incidence", data=df2)
plt.show()
```
